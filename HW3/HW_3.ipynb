#### 1.2. Store your data into data frame:



```python
# Code for storing data into data frame

#only 826 moves
import requests as rq
import pandas as pd

moveUrl ="https://pokeapi.co/api/v2/move/"


def getMove(n): #in pokemon each move has an unique number and this api lets us request moves by number
    r = rq.get(moveUrl+str(n)+'/')
    return(r)


power = {}
generation = {}
accuracy = {}



for n in range(1, 827): #there is only 826 moves 
    move = getMove(n)
    print(n)
    
    power[move.json()['name']] = move.json()['power']
    generation[move.json()['name']] = move.json()['generation']['name']
    accuracy[move.json()['name']] = move.json()['accuracy']


moves = pd.DataFrame({
    'power':power,
    'generation' : generation,
    'accuracy' : accuracy,
})

moves
```

# Cleaning the Data - not part of assignment but added since it makes my life easier

```python

#some Moves only inflict status effects and do not deal damange, hence there damage and accuracy is NaN. This may cause issues so I am removing all the NaN data values
moves = moves.dropna()

#Next I want to compare damage over each generation, but dont like that generation is stored as a stirng "generation-i" ect so I will turn it into an integer
index = 0
for n in moves.loc[:,'generation']: #this iterrates through everything in the generation coloumn
    if n == 'generation-i': #if generation = generation-i, replace it with 1 . . . the rest is just this repeated
        moves.iloc[index,1] = 1
    elif n == 'generation-ii':
        moves.iloc[index,1] = 2
    elif n == 'generation-iii':
        moves.iloc[index,1] = 3
    elif n == 'generation-iv':
        moves.iloc[index,1] = 4
    elif n == 'generation-v':
        moves.iloc[index,1] = 5
    elif n == 'generation-vi':
        moves.iloc[index,1] = 6
    elif n == 'generation-vii':
        moves.iloc[index,1] = 7
    elif n == 'generation-viii':
        moves.iloc[index,1] = 8
    index = index +1


moves
```

```python

# Code for basic statisitcs  

print("the mean power is "+str(moves.loc[:,'power'].mean()))
print("the mean accuracy is "+str(moves.loc[:,'accuracy'].mean()))
print("the mean generation is "+str(moves.loc[:,'generation'].mean()))

print("the max power is "+str(moves.loc[:,'power'].max()))
print("the max accuracy is "+str(moves.loc[:,'accuracy'].max()))
print("the max generation is "+str(moves.loc[:,'generation'].max()))

print("the min power is "+str(moves.loc[:,'power'].min()))
print("the min accuracy is "+str(moves.loc[:,'accuracy'].min()))
print("the min generation is "+str(moves.loc[:,'generation'].min()))

print("the standard deviation of power is "+str(moves.loc[:,'power'].std()))
print("the standard deviation of accuracy is "+str(moves.loc[:,'accuracy'].std()))
print("the standard generation of power is "+str(moves.loc[:,'generation'].std()))

print("the median power is "+str(moves.loc[:,'power'].median()))
print("the median accuracy is "+str(moves.loc[:,'accuracy'].median()))
print("the median generation is "+str(moves.loc[:,'generation'].median()))

print("the 25th percentile of power is "+str(moves.loc[:,'power'].quantile(0.25)))
print("the 25th percentile of accuracy is "+str(moves.loc[:,'accuracy'].quantile(0.25)))
print("the 25th percentile of generation is "+str(moves.loc[:,'generation'].quantile(0.25)))

print("the 75th percentile of power is "+str(moves.loc[:,'power'].quantile(0.75)))
print("the 75th percentile of accuracy is "+str(moves.loc[:,'accuracy'].quantile(0.75)))
print("the 75th percentile of generation is "+str(moves.loc[:,'generation'].quantile(0.75)))

moves
```
